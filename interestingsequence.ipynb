```python
'''
Sequence is defined as follows: 
 a_1 = 1
 a_n = n/a_{n-1}   , n >= 2
 
Seems to follow a sqrt shape when graphed. 
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
seq_recursive = [1.0]
seq2 = np.arange(1,75,1)
```

```python
# Recursive formula

for i in range(2,75):
    seq_recursive.append(i/seq_recursive[i-2])
```

```python
plt.plot(seq2,seq_recursive)
plt.plot(seq2, np.sqrt(seq2)*1.02)
plt.show()
```

```python
seq_exact = [1]
```

```python
# Exact formula

for n in range(2,101):
    new_iter = 1
    if n % 2 == 0:
        for i in range(1, n+1):
            new_iter *= i**((-1)**i)
    elif n % 2 == 1:
        for i in range(1, n+1):
            new_iter *= i**((-1)**(i-1))
    seq_exact.append(new_iter)
```

```python
plt.plot(seq_exact)
plt.show()
```

```python
# Split upper and lower subsequences

seq_evens = [seq_exact[i] for i in range(len(seq_exact)) if i % 2 == 0] # lower subsequence
seq_odds = [seq_exact[i] for i in range(len(seq_exact)) if i % 2 == 1] # upper subsequence
```

```python
evens = [i for i in range(100) if  i % 2 == 0] # even numbers
odds = [i for i in range(100) if i % 2 == 1] # odd numbers
```

```python
plt.scatter(evens, seq_evens)
plt.scatter(odds, seq_odds)

plt.show()
```
